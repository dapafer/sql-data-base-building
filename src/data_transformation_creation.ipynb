# **DATA TRANSFORMATION & CREATION.**

## **Libraries.**

```python
import numpy as np
import pandas as pd
import re
```

## **Data transformation.**

### **Import clean datasets.**

```python
actors = pd.read_csv('../data/actor_clean.csv')
categories = pd.read_csv('../data/category_clean.csv')
films = pd.read_csv('../data/film_clean.csv')
inventories = pd.read_csv('../data/inventory_clean.csv')
languages = pd.read_csv('../data/language_clean.csv')
hdd = pd.read_csv('../data/old_HDD_clean.csv')
rentals = pd.read_csv('../data/rental_clean.csv')
```

### **Actors-Films relation.**

```python
actors
```

```python
hdd
```

```python
films
```

**We will perform the fusion between "Actor" and "Film" dataframes through "old_HDD", since it contains common columns to be able to execute it.**

```python
merged_actors_hdd = pd.merge(actors, hdd, on=['first_name', 'last_name'])
merged_actors_hdd
```

```python
merged_actors_films = pd.merge(merged_actors_hdd, films, on='title', how='left')
merged_actors_films
```

**We will transform this dataframe to only keep the "actor_id" and "film_id" columns to make the relationship between the two dataframes in the database, and we create a new column called "actor_film_id".**

```python
merged_actors_films = merged_actors_films[['actor_id', 'film_id']]

actor_film_id = pd.Series(range(1, len(merged_actors_films) + 1))
merged_actors_films.insert(0, 'actor_film_id', actor_film_id)

merged_actors_films
```

```python
print('Duplicate values: ', merged_actors_films.duplicated().sum())
```

**We export the resulting dataframe for the relation in the database.**

```python
merged_actors_films.to_csv('../data/actor_film_clean.csv', index=False)
```

### **Categories-Films merge.**

```python
categories
```

```python
hdd
```

```python
films
```

**We also relate the two dataframes through "old_hdd", since they have columns in common.**

```python
merged_categories_hdd = pd.merge(hdd, categories, on='category_id', how='left')
merged_categories_hdd
```

```python
films = pd.merge(films, merged_categories_hdd, on='title', how='left')
films
```

```python
print('Duplicate values: ', films.duplicated().sum())
```

**Having incorporated the relationship through "category_id", we proceed to transform the dataframe and delete the columns that are not necessary, including null values.**

```python
films.drop('first_name', axis=1, inplace=True)
films.drop('last_name', axis=1, inplace=True)
films.drop('release_year_y', axis=1, inplace=True)
films.drop('category_id', axis=1, inplace=True)

films = films.rename(columns= {'release_year_x' : 'release_year'})
films = films.rename(columns= {'name' : 'category'})

films
```

```python
print('Duplicate values: ', films.duplicated().sum())
```

```python
films.drop_duplicates(inplace=True)
films
```

**We fill the null values with "unknown".**

```python
films['category'].fillna('unknown', inplace=True)
films
```

**Export the new dataframe for incorporate in the database.**

```python
films.to_csv('../data/film_clean.csv', index=False)
```
